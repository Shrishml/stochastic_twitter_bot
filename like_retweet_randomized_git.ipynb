```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.wait import WebDriverWait
import time
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys



a=np.random.normal(loc=15, scale=5.0, size=100)  # these are for creating randomization, Twitter AI will not throw captcha at you
b=np.random.normal(loc=6, scale=3.0, size=100)
c= np.random.normal(loc=17, scale=4.5, size=100)

# ID_PASS is the list of id and password of your profiles.

ID_PASS = [["abcd@gmail.com", "Password"],['vbdt@yahoo.com', 'PaSSword'], ['916542****', 'PASSWORD']]  # an example list

# handles are list of profile you want to like and retweet for (everythig after @)

handles=['JoeBiden','donaltrumparody','Wisdom'] # an example list


#print(handles[-4:-1])




for link in handles:
    for hand in ID_PASS:
        driver = webdriver.Chrome(executable_path="./chromedriver")
        driver.implicitly_wait(10)
        #driver.maximize_window()

        driver.get('https://twitter.com/login')
        time.sleep(3)
        driver.find_element_by_name('session[username_or_email]').send_keys(hand[0])
        driver.find_element_by_name('session[password]').send_keys(hand[1] + Keys.RETURN)
        #time.sleep(np.random.choice(b))
        driver.get("https://twitter.com/"+link)
        time.sleep(np.random.choice(c))
        #driver.execute_script('window.scrollTo(0,6000)')
        time.sleep(4)

        for i in range(70):
            if(i<13):
                actions = ActionChains(driver)
                actions.send_keys("j")
                time.sleep(np.random.choice(b))
                actions.send_keys("l")
                time.sleep(np.random.choice(b))
                actions.perform()

            if (i > 12):
                actions = ActionChains(driver)
                actions.send_keys("j")
                time.sleep(np.random.choice(b))
                actions.send_keys("l")
                time.sleep(np.random.choice(b))
                actions.perform()
                if(np.random.uniform()<0.7):
                    actions = ActionChains(driver)
                    actions.send_keys("t")
                    time.sleep(np.random.choice(b))
                    actions.send_keys(Keys.RETURN)
                    actions.perform()

            time.sleep(np.random.choice(a))

        driver.quit()

print("done")

```
